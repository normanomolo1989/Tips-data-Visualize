```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE
sm=SMOTE()
print(Counter (y_train))
X_sm,y_sm=sm.fit_resample(X_train,y_train)
print(Counter(y_sm))
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_sm_sc=sc.fit_transform(X_sm)
```

```python
X_sm_sc.shape[0]
```

```python
y_sm.shape
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(X_sm_sc,y_sm)
```

```python
y_sm_pre=lg.predict(X_sm_sc)
```

```python
y_sm_pre
```

```python
y_sm
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score
```

```python
acc=accuracy_score(y_sm,y_sm_pre) #model training accuracy
acc
```

```python
y_pred=lg.predict(X_test)#prediction on testing model
```

```python
acc=accuracy_score(y_test,y_pred)#model accuracy  on testing data
acc
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
y_tr_pred=knn.predict(X_sm_sc)#performance on training model
```

```python
y_tr_pred
```

```python
acc=accuracy_score(y_sm,y_tr_pred)
acc
```
